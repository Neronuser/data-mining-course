## 1. Задача класифікації в data mining

### Що таке класифікація?
Класифікація - це задача навчання з вчителем, де ми прогнозуємо категоріальні мітки (класи) для нових екземплярів на основі патернів, вивчених з розмічених навчальних даних. Це одна з найфундаментальніших проблем у інтелектуальному аналізі даних та машинному навчанні.

**Ключові характеристики:**
- **Вхід**: Набір даних з ознаками та відомими мітками класів
- **Вихід**: Модель, яка може прогнозувати клас нових екземплярів
- **Мета**: Вивчити межі рішень, які розділяють різні класи

### Застосування у реальному світі
- **Медична діагностика**: Класифікувати пацієнтів як здорових або хворих на основі симптомів
- **Виявлення спаму в електронній пошті**: Класифікувати електронні листи як спам або не спам
- **Оцінка кредитного ризику**: Класифікувати заявників на кредит як кандидатів з високим або низьким ризиком
- **Розпізнавання зображень**: Класифікувати зображення на категорії (кіт, собака, птах тощо)

### Класифікація vs. Інші задачі ІАД
- **Класифікація**: Прогнозування категоріальних міток (дискретні класи)
- **Регресія**: Прогнозування неперервних числових значень
- **Кластеризація**: Знаходження прихованих патернів без розмічених даних (навчання без вчителя)
- **Асоціативні правила**: Знаходження зв'язків між елементами (аналіз кошика покупок)

## 2. Дерева рішень

### Що таке дерева рішень?
Дерева рішень - це популярний та інтерпретований алгоритм класифікації, який представляє рішення та їх можливі наслідки у деревоподібній структурі. Вони імітують процеси прийняття рішень людиною, ставлячи серію питань про ознаки.

**Ключові переваги методу:**
- **Інтерпретований**: Легко зрозуміти та пояснити
- **Без припущень**: Не вимагає, щоб дані слідували конкретним розподілам
- **Обробляє змішані дані**: Працює як з числовими, так і з категоріальними ознаками
- **Вибір ознак**: Автоматично ідентифікують важливі ознаки
- **Будує правила**: Може бути перетворений на правила if-then

### Структура дерева
Дерево рішень має:
- **Кореневий вузол**: Найвищий вузол, який містить всі навчальні дані
- **Внутрішні вузли**: Вузли рішень, які перевіряють ознаку
- **Листові вузли**: Термінальні вузли, які містять фінальну класифікацію
- **Гілки**: Шляхи від кореня до листків, що представляють правила рішень

### Алгоритми дерев рішень

### Процес побудови дерева рішень
Дерево рішень будується, використовуючи підхід **рекурсивного розбиття**:

1. **Початок**: Почати з кореневого вузла, що містить всі навчальні дані
2. **Вибрати найкраще розбиття**: Вибрати ознаку, яка максимізує інформаційний приріст (або мінімізує Gini-impurity)
3. **Створити гілки**: Розбити дані на основі вибраної ознаки
4. **Рекурсія**: Повторити процес для кожного дочірнього вузла
5. **Зупинитися**: Зупинитися, коли виконано критерій зупинки

### Критерії зупинки
- **Чистий вузол**: Всі екземпляри у вузлі належать до одного класу
- **Немає ознак**: Більше немає доступних ознак для розбиття
- **Мінімальна кількість зразків**: Вузол має менше, ніж мінімальну необхідну кількість зразків
- **Максимальна глибина**: Дерево досягло максимально дозволеної глибини
- **Мінімальний інформаційний приріст**: Жодне розбиття не надає достатнього інформаційного приросту

#### ID3 (Iterative Dichotomiser 3)
- Використовує **ентропію** та **інформаційний приріст** для розбиття
- Обробляє лише категоріальні ознаки
- Не має механізму відсікання
- Може легко перенавчатись

#### C4.5 (Наступник ID3)
- Розширення ID3 з покращеннями
- Обробляє як категоріальні, так і числові ознаки
- Включає **відсікання** для зменшення перенавчання
- Використовує **коефіцієнт приросту** замість інформаційного приросту

#### CART (Classification and Regression Trees)
- Використовує **Gini-impurity** для розбиття
- Обробляє як категоріальні, так і числові ознаки
- Лише двійкові розбиття (створює двійкові дерева)
- Включає механізми відсікання

## 3. Ентропія та інформаційний приріст

### Ентропія
Ентропія - це міра невизначеності в наборі даних. У контексті дерев рішень ми хочемо мінімізувати ентропію, щоб створити чисті (однорідні) вузли.

**Математичне визначення:**
$$\text{Ентропія}(S) = -\sum_{i=1}^{c} p_i \log_2(p_i)$$

Де:
- $S$ - набір даних
- $c$ - кількість класів
- $p_i$ - частка екземплярів, що належать до класу $i$

### Властивості ентропії
- **Мінімум (0)**: Коли всі екземпляри належать до одного класу (чистий вузол)
- **Максимум (log₂c)**: Коли всі класи представлені однаково (максимальна невизначеність)
- **Основа логарифма 2**: Дає ентропію в бітах

### Інформаційний приріст
Інформаційний приріст вимірює зменшення ентропії, досягнуте розбиттям за конкретною ознакою. Ми вибираємо ознаку, яка надає максимальний інформаційний приріст.

**Математичне визначення:**
$$\text{Інформаційний приріст}(S, A) = \text{Ентропія}(S) - \sum_{v \in \text{Значення}(A)} \frac{|S_v|}{|S|} \text{Ентропія}(S_v)$$

Де:
- $S$ - оригінальний набір даних
- $A$ - ознака, що тестується
- $S_v$ - підмножина $S$, де ознака $A$ має значення $v$
- $|S_v|$ - розмір підмножини $S_v$
- $|S|$ - розмір оригінального набору даних $S$

### Приклад: Обчислення ентропії та інформаційного приросту
Давайте пройдемо через простий приклад з малим набором даних, щоб зрозуміти ці концепції.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

def entropy(labels):
    label_counts = Counter(labels)
    total = len(labels)
    entropy_value = 0
    for count in label_counts.values():
        p = count / total
        if p > 0:
            entropy_value -= p * np.log2(p)
    return entropy_value

def information_gain(data, feature, target):
    """
    Обчислює інформаційний приріст для ознаки
    
    Параметри:
        data: DataFrame з ознаками та цільовою змінною
        feature: Назва ознаки для тестування
        target: Назва стовпця цільової змінної
    
    Повертає:
        Значення інформаційного приросту
    """
    original_entropy = entropy(data[target])
    weighted_entropy = 0
    total_samples = len(data)
    for value in data[feature].unique():
        subset = data[data[feature] == value]
        subset_size = len(subset)
        subset_entropy = entropy(subset[target])
        weighted_entropy += (subset_size / total_samples) * subset_entropy
    return original_entropy - weighted_entropy

print("=== Простий приклад: Набір даних 'Грати в теніс' ===")
print()

tennis_data = pd.DataFrame({
    'Погода': ['Сонячно', 'Сонячно', 'Похмуро', 'Дощ', 'Дощ', 'Дощ', 'Похмуро', 
                'Сонячно', 'Сонячно', 'Дощ', 'Сонячно', 'Похмуро', 'Похмуро', 'Дощ'],
    'Температура': ['Жарко', 'Жарко', 'Жарко', 'Помірно', 'Прохолодно', 'Прохолодно', 'Прохолодно', 
                    'Помірно', 'Прохолодно', 'Помірно', 'Помірно', 'Помірно', 'Жарко', 'Помірно'],
    'Вологість': ['Висока', 'Висока', 'Висока', 'Висока', 'Нормальна', 'Нормальна', 'Нормальна', 
                 'Висока', 'Нормальна', 'Нормальна', 'Нормальна', 'Висока', 'Нормальна', 'Висока'],
    'Вітер': ['Слабкий', 'Сильний', 'Слабкий', 'Слабкий', 'Слабкий', 'Сильний', 'Сильний', 
             'Слабкий', 'Слабкий', 'Слабкий', 'Сильний', 'Сильний', 'Слабкий', 'Сильний'],
    'Грати в теніс': ['Ні', 'Ні', 'Так', 'Так', 'Так', 'Ні', 'Так', 
                   'Ні', 'Так', 'Так', 'Так', 'Так', 'Так', 'Ні']
})

print("Набір даних:")
print(tennis_data)
print()

target_entropy = entropy(tennis_data['Грати в теніс'])
print(f"Ентропія всього набору даних: {target_entropy:.3f}")

features = ['Погода', 'Температура', 'Вологість', 'Вітер']
print("\nІнформаційний приріст для кожної ознаки:")
print("-" * 40)

for feature in features:
    ig = information_gain(tennis_data, feature, 'Грати в теніс')
    print(f"{feature:12}: {ig:.3f}")

print(f"\nНайкраща ознака для розбиття: {max(features, key=lambda f: information_gain(tennis_data, f, 'Грати в теніс'))}")
````

## 4. Gini-impurity: Альтернативний критерій розбиття

### Що таке Gini-impurity?
Gini-impurity - це інша міра невизначеності, яку використовують в деревах рішень, особливо в алгоритмі CART. Вона вимірює ймовірність неправильної класифікації випадково обраного елемента, якщо він був би випадково позначений відповідно до розподілу міток у підмножині.

**Математичне визначення:**
$$\text{Gini}(S) = 1 - \sum_{i=1}^{c} p_i^2$$

Де:
- $S$ - набір даних
- $c$ - кількість класів
- $p_i$ - частка екземплярів, що належать до класу $i$

### Властивості Gini-impurity
- **Мінімум (0)**: Коли всі екземпляри належать до одного класу (чистий вузол)
- **Максимум (1 - 1/c)**: Коли всі класи представлені однаково
- **Діапазон**: [0, 1] для двійкової класифікації, [0, 1-1/c] для багатокласової

### Gini vs. Ентропія
- **Gini**: Швидше обчислюється (немає логарифма)
- **Ентропія**: Більш чутлива до малих змін у ймовірності
- **Подібна поведінка**: Обидві досягають мінімуму при чистих вузлах та максимуму при рівномірному розподілі
- **CART використовує Gini**: Більш обчислювально ефективна для великих наборів даних

### Приріст Gini
Подібно до інформаційного приросту, ми обчислюємо приріст Gini, щоб визначити найкращу ознаку для розбиття:

$$\text{Приріст Gini}(S, A) = \text{Gini}(S) - \sum_{v \in \text{Значення}(A)} \frac{|S_v|}{|S|} \text{Gini}(S_v)$$

````python

def gini_impurity(labels):
    label_counts = Counter(labels)
    total = len(labels)
    gini = 1
    for count in label_counts.values():
        p = count / total
        gini -= p ** 2
    return gini

def gini_gain(data, feature, target):
    original_gini = gini_impurity(data[target])
    weighted_gini = 0
    total_samples = len(data)
    for value in data[feature].unique():
        subset = data[data[feature] == value]
        subset_size = len(subset)
        subset_gini = gini_impurity(subset[target])
        weighted_gini += (subset_size / total_samples) * subset_gini
    return original_gini - weighted_gini

print("=== Порівняння ентропії та Gini-impurity ===")
print()

target_entropy = entropy(tennis_data['Грати в теніс'])
target_gini = gini_impurity(tennis_data['Грати в теніс'])

print(f"Показники цільової змінної:")
print(f"   Ентропія: {target_entropy:.3f}")
print(f"   Gini:     {target_gini:.3f}")
print()

print("Порівняння ознак (Ентропія vs Gini):")
print("-" * 50)
print(f"{'Ознака':<12} {'Інф. приріст':<12} {'Gini приріст':<12} {'Різниця':<12}")
print("-" * 50)

for feature in features:
    ig = information_gain(tennis_data, feature, 'Грати в теніс')
    gg = gini_gain(tennis_data, feature, 'Грати в теніс')
    diff = abs(ig - gg)
    print(f"{feature:<12} {ig:<12.3f} {gg:<12.3f} {diff:<12.3f}")

print(f"\nНайкраща ознака (Інф. приріст): {max(features, key=lambda f: information_gain(tennis_data, f, 'Грати в теніс'))}")
print(f"Найкраща ознака (Gini приріст): {max(features, key=lambda f: gini_gain(tennis_data, f, 'Грати в теніс'))}")

print("\nВізуалізація залежності ентропії і Gini-impurity...")

probs = np.linspace(0.01, 0.99, 100)
entropies = [-p * np.log2(p) - (1-p) * np.log2(1-p) for p in probs]
ginis = [1 - p**2 - (1-p)**2 for p in probs]

plt.figure(figsize=(10, 6))
plt.plot(probs, entropies, 'b-', label='Ентропія', linewidth=2)
plt.plot(probs, ginis, 'r--', label='Індекс Gini', linewidth=2)
plt.xlabel('Ймовірність класу 1')
plt.ylabel('Міра невизначеності')
plt.title('Ентропія vs індекс Gini для бінарної класифікації')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## 5. Відсікання (Pruning)

### Проблема перенавчання
Дерева рішень можуть легко **перенавчатись** - запамʼятовувати навчальні дані, створюючи надто складні дерева, які не узагальнюються до нових даних.

**Ознаки перенавчання:**
- Дуже глибокі дерева з багатьма рівнями
- Багато листових вузлів з дуже малою кількістю зразків
- Високі метрики на навчальних даних, але низькі метрики на тестових даних
- Дерева, які запам'ятовують навчальні приклади, а не вивчають патерни

### Відсікання: Рішення проблеми перенавчання
Відсікання - це процес видалення гілок з дерева рішень для покращення узагальнення.

#### Попереднє Відсікання (Рання зупинка)
Зупинити побудову дерева до того, як воно стане надто складним:
- Встановити максимальну глибину
- Встановити мінімальну кількість зразків на лист
- Встановити поріг мінімального інформаційного приросту

### Переваги Відсікання
- **Боротьба з перенавчанням**: Краще узагальнення до нових даних
- **Простіші моделі**: Легше зрозуміти та інтерпретувати
- **Швидше прогнозування**: Менше рішень для прийняття
- **Краща точність**: Часто покращує точність на тестових даних

## 6. Інтерпретованість

Однією з головних переваг дерев рішень є їх **інтерпретованість**. Ми можемо легко перетворити дерево рішень на набір правил if-then, які є:
- **Зрозумілими для людини**: Легко зрозуміти та пояснити
- **Дієвими**: Можуть бути реалізовані в бізнес-логіці
- **Відлагоджуваними**: Легко відстежити та валідувати рішення
- **Портативними**: Можуть використовуватися в різних системах та мовах

### Формат правила
```
ЯКЩО умова1 І умова2 І ... І умоваN 
ТОДІ клас = прогнозований_клас
```

### Приклад: Перетворення дерева "Грати в теніс" на правила
Давайте подивимося, як наше дерево рішень для тенісу виглядало б як правила:

**Правило 1**: `ЯКЩО Погода = Похмуро ТОДІ Грати в теніс = Так`

**Правило 2**: `ЯКЩО Погода = Сонячно І Вологість = Висока ТОДІ Грати в теніс = Ні`

**Правило 3**: `ЯКЩО Погода = Сонячно І Вологість = Нормальна ТОДІ Грати в теніс = Так`

**Правило 4**: `ЯКЩО Погода = Дощ І Вітер = Strong ТОДІ Грати в теніс = Ні`

**Правило 5**: `ЯКЩО Погода = Дощ І Вітер = Weak ТОДІ Грати в теніс = Так`

### Метрики якості правил
- **Покриття**: Скільки навчальних прикладів покриває правило
- **Точність**: Скільки покритих прикладів правильно класифіковано
- **Підтримка**: Кількість прикладів, які задовольняють правило
- **Впевненість**: Частка правильно класифікованих прикладів серед покритих

### Спрощення правил
Правила часто можна спростити:
- **Видалення надлишкових умов**: Якщо умова не впливає на результат
- **Поєднання подібних правил**: Об'єднати правила з подібними умовами
- **Впорядкування за важливістю**: Сортувати правила за впевненістю або покриттям

## 7. Практична реалізація з Scikit-Learn

Набір даних Iris містить виміри квітів ірисів з трьома видами:
- **Setosa**: 50 зразків
- **Versicolor**: 50 зразків

**Ознаки:**
- Довжина чашолистка (см)

- Довжина пелюстки (см)

````python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

print("=== Завантаження набору даних Iris ===")
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names
target_names = iris.target_names

print(f"Форма набору даних: {X.shape}")
print(f"Ознаки: {feature_names}")
print(f" Класи: {target_names}")
print(f"Розподіл по класах: {np.bincount(y)}")
print()

iris_df = pd.DataFrame(X, columns=feature_names)
iris_df['species'] = [target_names[i] for i in y]

print("Перші кілька рядків набору даних:")
print(iris_df.head())
print()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Розподіл даних:")
print(f"   Навчальна вибірка: {X_train.shape[0]} зразків")
print(f"   Тестова вибірка:  {X_test.shape[0]} зразків")
print()

print("Навчання дерева рішень...")
dt_classifier = DecisionTreeClassifier(
    criterion='gini',
    max_depth=3,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)

dt_classifier.fit(X_train, y_train)

y_pred = dt_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Точність на тестових даних: {accuracy:.3f}")
print()

print("Класифікаційний звіт:")
print(classification_report(y_test, y_pred, target_names=target_names))
````

````python
plt.figure(figsize=(15, 10))
plot_tree(dt_classifier, 
          feature_names=feature_names,
          class_names=target_names,
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Дерево рішень для класифікації Iris", fontsize=16, fontweight='bold')
plt.show()

print("Правила дерева рішень (текстовий формат):")
print("=" * 50)
tree_rules = export_text(dt_classifier, 
                        feature_names=feature_names,
                        class_names=target_names,
                        show_weights=True)
print(tree_rules)

print("\nАналіз важливості ознак:")
print("-" * 40)
feature_importance = dt_classifier.feature_importances_
for i, (feature, importance) in enumerate(zip(feature_names, feature_importance)):
    print(f"{feature:20}: {importance:.3f}")

plt.figure(figsize=(10, 6))
plt.barh(feature_names, feature_importance)
plt.xlabel('Важливість ознаки')
plt.title('Важливість ознак у дереві рішень')
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nНайважливіша ознака: {feature_names[np.argmax(feature_importance)]}")
````

## 8. Переваги та недоліки дерев рішень

### Переваги

#### Інтерпретованість
- **Зрозумілі для людини**: Легко зрозуміти та пояснити нетехнічним стейкхолдерам
- **Візуальне представлення**: Структура дерева надає чіткі шляхи рішень
- **Побудова правил**: Можуть бути перетворені на правила if-then для бізнес-логіки

#### Гнучкість
- **Змішані типи даних**: Обробляють як числові, так і категоріальні ознаки
- **Без припущень**: Не вимагають, щоб дані слідували конкретним розподілам
- **Вибір ознак**: Автоматично ідентифікують важливі ознаки
- **Відсутні значення**: Можуть обробляти відсутні дані

#### Ефективність
- **Швидке навчання**: Відносно швидко будувати порівняно з іншими алгоритмами
- **Швидке прогнозування**: Дуже швидка класифікація нових екземплярів
- **Ефективність пам'яті**: Не вимагають збереження всього набору даних після навчання

### Недоліки

#### Перенавчання
- **Схильність до перенавчання**: Можуть створювати надто складні дерева, які запам'ятовують навчальні дані
- **Чутливість до шуму**: Малі зміни в даних можуть призвести до дуже різних дерев
- **Вимагає відсікання**: Потрібне ретельне налаштування параметрів для запобігання перенавчанню

#### Нестабільність
- **Висока дисперсія**: Малі зміни в навчальних даних можуть призвести до абсолютно різних дерев
- **Жадібний підхід**: Приймає локально оптимальні рішення, які можуть не бути глобально оптимальними
- **Немає глобальної оптимізації**: Не враховує загальну структуру дерева під час побудови

#### Обмеження
- **Розбиття, вирівняні по осях**: Може створювати лише розбиття, паралельні осям ознак
- **Погано з лінійними зв'язками**: Має проблеми з лінійними межами рішень
- **Незбалансовані дані**: Може створювати упереджені дерева, коли класи незбалансовані
