–ï–Ω—Ç—Ä–æ–ø—ñ—è - —Ü–µ –º—ñ—Ä–∞ –Ω–µ–≤–∏–∑–Ω–∞—á–µ–Ω–æ—Å—Ç—ñ –≤ –Ω–∞–±–æ—Ä—ñ –¥–∞–Ω–∏—Ö. –£ –∫–æ–Ω—Ç–µ–∫—Å—Ç—ñ –¥–µ—Ä–µ–≤ —Ä—ñ—à–µ–Ω—å –º–∏ —Ö–æ—á–µ–º–æ –º—ñ–Ω—ñ–º—ñ–∑—É–≤–∞—Ç–∏ –µ–Ω—Ç—Ä–æ–ø—ñ—é, —â–æ–± —Å—Ç–≤–æ—Ä–∏—Ç–∏ —á–∏—Å—Ç—ñ (–æ–¥–Ω–æ—Ä—ñ–¥–Ω—ñ) –≤—É–∑–ª–∏.

**–ú–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è:**
$$\text{–ï–Ω—Ç—Ä–æ–ø—ñ—è}(S) = -\sum_{i=1}^{c} p_i \log_2(p_i)$$

–î–µ:
- $S$ - –Ω–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö
- $c$ - –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—ñ–≤
- $p_i$ - —á–∞—Å—Ç–∫–∞ –µ–∫–∑–µ–º–ø–ª—è—Ä—ñ–≤, —â–æ –Ω–∞–ª–µ–∂–∞—Ç—å –¥–æ –∫–ª–∞—Å—É $i$

### –í–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ –µ–Ω—Ç—Ä–æ–ø—ñ—ó
- **–ú—ñ–Ω—ñ–º—É–º (0)**: –ö–æ–ª–∏ –≤—Å—ñ –µ–∫–∑–µ–º–ø–ª—è—Ä–∏ –Ω–∞–ª–µ–∂–∞—Ç—å –¥–æ –æ–¥–Ω–æ–≥–æ –∫–ª–∞—Å—É (—á–∏—Å—Ç–∏–π –≤—É–∑–æ–ª)
- **–ú–∞–∫—Å–∏–º—É–º (log‚ÇÇc)**: –ö–æ–ª–∏ –≤—Å—ñ –∫–ª–∞—Å–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ñ –æ–¥–Ω–∞–∫–æ–≤–æ (–º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞ –Ω–µ–≤–∏–∑–Ω–∞—á–µ–Ω—ñ—Å—Ç—å)
- **–û—Å–Ω–æ–≤–∞ –ª–æ–≥–∞—Ä–∏—Ñ–º–∞ 2**: –î–∞—î –µ–Ω—Ç—Ä–æ–ø—ñ—é –≤ –±—ñ—Ç–∞—Ö

### –Ü–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–∏–π –ø—Ä–∏—Ä—ñ—Å—Ç
–Ü–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–∏–π –ø—Ä–∏—Ä—ñ—Å—Ç –≤–∏–º—ñ—Ä—é—î –∑–º–µ–Ω—à–µ–Ω–Ω—è –µ–Ω—Ç—Ä–æ–ø—ñ—ó, –¥–æ—Å—è–≥–Ω—É—Ç–µ —Ä–æ–∑–±–∏—Ç—Ç—è–º –∑–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ—é –æ–∑–Ω–∞–∫–æ—é. –ú–∏ –≤–∏–±–∏—Ä–∞—î–º–æ –æ–∑–Ω–∞–∫—É, —è–∫–∞ –Ω–∞–¥–∞—î –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–∏–π –ø—Ä–∏—Ä—ñ—Å—Ç.

**–ú–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è:**
$$\text{–Ü–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–∏–π –ø—Ä–∏—Ä—ñ—Å—Ç}(S, A) = \text{–ï–Ω—Ç—Ä–æ–ø—ñ—è}(S) - \sum_{v \in \text{–ó–Ω–∞—á–µ–Ω–Ω—è}(A)} \frac{|S_v|}{|S|} \text{–ï–Ω—Ç—Ä–æ–ø—ñ—è}(S_v)$$

–î–µ:
- $S$ - –æ—Ä–∏–≥—ñ–Ω–∞–ª—å–Ω–∏–π –Ω–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö
- $A$ - –æ–∑–Ω–∞–∫–∞, —â–æ —Ç–µ—Å—Ç—É—î—Ç—å—Å—è
- $S_v$ - –ø—ñ–¥–º–Ω–æ–∂–∏–Ω–∞ $S$, –¥–µ –æ–∑–Ω–∞–∫–∞ $A$ –º–∞—î –∑–Ω–∞—á–µ–Ω–Ω—è $v$
- $|S_v|$ - —Ä–æ–∑–º—ñ—Ä –ø—ñ–¥–º–Ω–æ–∂–∏–Ω–∏ $S_v$
- $|S|$ - —Ä–æ–∑–º—ñ—Ä –æ—Ä–∏–≥—ñ–Ω–∞–ª—å–Ω–æ–≥–æ –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö $S$

### –ü—Ä–∏–∫–ª–∞–¥: –û–±—á–∏—Å–ª–µ–Ω–Ω—è –µ–Ω—Ç—Ä–æ–ø—ñ—ó —Ç–∞ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–æ–≥–æ –ø—Ä–∏—Ä–æ—Å—Ç—É
–î–∞–≤–∞–π—Ç–µ –ø—Ä–æ–π–¥–µ–º–æ —á–µ—Ä–µ–∑ –ø—Ä–æ—Å—Ç–∏–π –ø—Ä–∏–∫–ª–∞–¥ –∑ –º–∞–ª–∏–º –Ω–∞–±–æ—Ä–æ–º –¥–∞–Ω–∏—Ö, —â–æ–± –∑—Ä–æ–∑—É–º—ñ—Ç–∏ —Ü—ñ –∫–æ–Ω—Ü–µ–ø—Ü—ñ—ó.

````python
# Let's implement entropy and information gain calculations step by step!

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

def entropy(labels):
    """
    Calculate entropy for a set of labels
    
    Args:
        labels: List or array of class labels
    
    Returns:
        Entropy value
    """
    # Count occurrences of each class
    label_counts = Counter(labels)
    total = len(labels)
    
    # Calculate entropy
    entropy_value = 0
    for count in label_counts.values():
        p = count / total  # proportion of this class
        if p > 0:  # avoid log(0)
            entropy_value -= p * np.log2(p)
    
    return entropy_value

def information_gain(data, feature, target):
    """
    Calculate information gain for a feature
    
    Args:
        data: DataFrame with features and target
        feature: Name of the feature to test
        target: Name of the target column
    
    Returns:
        Information gain value
    """
    # Calculate entropy of the original dataset
    original_entropy = entropy(data[target])
    
    # Calculate weighted entropy after splitting
    weighted_entropy = 0
    total_samples = len(data)
    
    for value in data[feature].unique():
        subset = data[data[feature] == value]
        subset_size = len(subset)
        subset_entropy = entropy(subset[target])
        weighted_entropy += (subset_size / total_samples) * subset_entropy
    
    # Information gain = original entropy - weighted entropy
    return original_entropy - weighted_entropy

# Example: Simple dataset for understanding entropy and information gain
print("=== Simple Example: Play Tennis Dataset ===")
print()

# Create a simple dataset
tennis_data = pd.DataFrame({
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 
                'Sunny', 'Sunny', 'Rain', 'Sunny', 'Overcast', 'Overcast', 'Rain'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 
                    'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 
                 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 
             'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Strong'],
    'PlayTennis': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 
                   'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
})

print("Dataset:")
print(tennis_data)
print()

# Calculate entropy of the target variable
target_entropy = entropy(tennis_data['PlayTennis'])
print(f"Entropy of PlayTennis: {target_entropy:.3f}")

# Calculate information gain for each feature
features = ['Outlook', 'Temperature', 'Humidity', 'Wind']
print("\nInformation Gain for each feature:")
print("-" * 40)

for feature in features:
    ig = information_gain(tennis_data, feature, 'PlayTennis')
    print(f"{feature:12}: {ig:.3f}")

print(f"\nBest feature to split on: {max(features, key=lambda f: information_gain(tennis_data, f, 'PlayTennis'))}")
````

## 4. Gini-impurity: –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä—ñ–π —Ä–æ–∑–±–∏—Ç—Ç—è

### –©–æ —Ç–∞–∫–µ Gini-impurity?
Gini-impurity - —Ü–µ —ñ–Ω—à–∞ –º—ñ—Ä–∞ –Ω–µ–≤–∏–∑–Ω–∞—á–µ–Ω–æ—Å—Ç—ñ, —è–∫—É –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å –≤ –¥–µ—Ä–µ–≤–∞—Ö —Ä—ñ—à–µ–Ω—å, –æ—Å–æ–±–ª–∏–≤–æ –≤ –∞–ª–≥–æ—Ä–∏—Ç–º—ñ CART. –í–æ–Ω–∞ –≤–∏–º—ñ—Ä—é—î –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ—ó –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó –≤–∏–ø–∞–¥–∫–æ–≤–æ –æ–±—Ä–∞–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞, —è–∫—â–æ –≤—ñ–Ω –±—É–≤ –±–∏ –≤–∏–ø–∞–¥–∫–æ–≤–æ –ø–æ–∑–Ω–∞—á–µ–Ω–∏–π –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ —Ä–æ–∑–ø–æ–¥—ñ–ª—É –º—ñ—Ç–æ–∫ —É –ø—ñ–¥–º–Ω–æ–∂–∏–Ω—ñ.

**–ú–∞—Ç–µ–º–∞—Ç–∏—á–Ω–µ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è:**
$$\text{Gini}(S) = 1 - \sum_{i=1}^{c} p_i^2$$

–î–µ:
- $S$ - –Ω–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö
- $c$ - –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—ñ–≤
- $p_i$ - —á–∞—Å—Ç–∫–∞ –µ–∫–∑–µ–º–ø–ª—è—Ä—ñ–≤, —â–æ –Ω–∞–ª–µ–∂–∞—Ç—å –¥–æ –∫–ª–∞—Å—É $i$

### –í–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ Gini-impurity
- **–ú—ñ–Ω—ñ–º—É–º (0)**: –ö–æ–ª–∏ –≤—Å—ñ –µ–∫–∑–µ–º–ø–ª—è—Ä–∏ –Ω–∞–ª–µ–∂–∞—Ç—å –¥–æ –æ–¥–Ω–æ–≥–æ –∫–ª–∞—Å—É (—á–∏—Å—Ç–∏–π –≤—É–∑–æ–ª)
- **–ú–∞–∫—Å–∏–º—É–º (1 - 1/c)**: –ö–æ–ª–∏ –≤—Å—ñ –∫–ª–∞—Å–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ñ –æ–¥–Ω–∞–∫–æ–≤–æ
- **–î—ñ–∞–ø–∞–∑–æ–Ω**: [0, 1] –¥–ª—è –¥–≤—ñ–π–∫–æ–≤–æ—ó –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó, [0, 1-1/c] –¥–ª—è –±–∞–≥–∞—Ç–æ–∫–ª–∞—Å–æ–≤–æ—ó

### Gini vs. –ï–Ω—Ç—Ä–æ–ø—ñ—è
- **Gini**: –®–≤–∏–¥—à–µ –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è (–Ω–µ–º–∞—î –ª–æ–≥–∞—Ä–∏—Ñ–º–∞)
- **–ï–Ω—Ç—Ä–æ–ø—ñ—è**: –ë—ñ–ª—å—à —á—É—Ç–ª–∏–≤–∞ –¥–æ –º–∞–ª–∏—Ö –∑–º—ñ–Ω —É –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ
- **–ü–æ–¥—ñ–±–Ω–∞ –ø–æ–≤–µ–¥—ñ–Ω–∫–∞**: –û–±–∏–¥–≤—ñ –¥–æ—Å—è–≥–∞—é—Ç—å –º—ñ–Ω—ñ–º—É–º—É –ø—Ä–∏ —á–∏—Å—Ç–∏—Ö –≤—É–∑–ª–∞—Ö —Ç–∞ –º–∞–∫—Å–∏–º—É–º—É –ø—Ä–∏ —Ä—ñ–≤–Ω–æ–º—ñ—Ä–Ω–æ–º—É —Ä–æ–∑–ø–æ–¥—ñ–ª—ñ
- **CART –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î Gini**: –ë—ñ–ª—å—à –æ–±—á–∏—Å–ª—é–≤–∞–ª—å–Ω–æ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–∞ –¥–ª—è –≤–µ–ª–∏–∫–∏—Ö –Ω–∞–±–æ—Ä—ñ–≤ –¥–∞–Ω–∏—Ö

### –ü—Ä–∏—Ä—ñ—Å—Ç Gini
–ü–æ–¥—ñ–±–Ω–æ –¥–æ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ–π–Ω–æ–≥–æ –ø—Ä–∏—Ä–æ—Å—Ç—É, –º–∏ –æ–±—á–∏—Å–ª—é—î–º–æ –ø—Ä–∏—Ä—ñ—Å—Ç Gini, —â–æ–± –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –Ω–∞–π–∫—Ä–∞—â—É –æ–∑–Ω–∞–∫—É –¥–ª—è —Ä–æ–∑–±–∏—Ç—Ç—è:

$$\text{–ü—Ä–∏—Ä—ñ—Å—Ç Gini}(S, A) = \text{Gini}(S) - \sum_{v \in \text{–ó–Ω–∞—á–µ–Ω–Ω—è}(A)} \frac{|S_v|}{|S|} \text{Gini}(S_v)$$

````python
# Let's implement Gini impurity and compare it with entropy!

def gini_impurity(labels):
    """
    Calculate Gini impurity for a set of labels
    
    Args:
        labels: List or array of class labels
    
    Returns:
        Gini impurity value
    """
    # Count occurrences of each class
    label_counts = Counter(labels)
    total = len(labels)
    
    # Calculate Gini impurity
    gini = 1
    for count in label_counts.values():
        p = count / total  # proportion of this class
        gini -= p ** 2
    
    return gini

def gini_gain(data, feature, target):
    """
    Calculate Gini gain for a feature
    
    Args:
        data: DataFrame with features and target
        feature: Name of the feature to test
        target: Name of the target column
    
    Returns:
        Gini gain value
    """
    # Calculate Gini impurity of the original dataset
    original_gini = gini_impurity(data[target])
    
    # Calculate weighted Gini impurity after splitting
    weighted_gini = 0
    total_samples = len(data)
    
    for value in data[feature].unique():
        subset = data[data[feature] == value]
        subset_size = len(subset)
        subset_gini = gini_impurity(subset[target])
        weighted_gini += (subset_size / total_samples) * subset_gini
    
    # Gini gain = original Gini - weighted Gini
    return original_gini - weighted_gini

# Let's compare entropy and Gini impurity on our tennis dataset
print("=== Comparing Entropy vs Gini Impurity ===")
print()

# Calculate both measures for the target variable
target_entropy = entropy(tennis_data['PlayTennis'])
target_gini = gini_impurity(tennis_data['PlayTennis'])

print(f"Target variable measures:")
print(f"   Entropy: {target_entropy:.3f}")
print(f"   Gini:    {target_gini:.3f}")
print()

# Compare information gain vs Gini gain for each feature
print("Feature comparison (Entropy vs Gini):")
print("-" * 50)
print(f"{'Feature':<12} {'Info Gain':<12} {'Gini Gain':<12} {'Difference':<12}")
print("-" * 50)

for feature in features:
    ig = information_gain(tennis_data, feature, 'PlayTennis')
    gg = gini_gain(tennis_data, feature, 'PlayTennis')
    diff = abs(ig - gg)
    print(f"{feature:<12} {ig:<12.3f} {gg:<12.3f} {diff:<12.3f}")

print(f"\nBest feature (Info Gain): {max(features, key=lambda f: information_gain(tennis_data, f, 'PlayTennis'))}")
print(f"Best feature (Gini Gain): {max(features, key=lambda f: gini_gain(tennis_data, f, 'PlayTennis'))}")

# Visualize the relationship between entropy and Gini
print("\nVisualizing Entropy vs Gini relationship...")

# Create a range of probabilities for binary classification
probs = np.linspace(0.01, 0.99, 100)
entropies = [-p * np.log2(p) - (1-p) * np.log2(1-p) for p in probs]
ginis = [1 - p**2 - (1-p)**2 for p in probs]

plt.figure(figsize=(10, 6))
plt.plot(probs, entropies, 'b-', label='Entropy', linewidth=2)
plt.plot(probs, ginis, 'r--', label='Gini Impurity', linewidth=2)
plt.xlabel('Probability of Class 1')
plt.ylabel('Impurity Measure')
plt.title('Entropy vs Gini Impurity for Binary Classification')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Notice how both measures have similar shapes and reach maximum at p=0.5!")
````

## 5. –í—ñ–¥—Å—ñ–∫–∞–Ω–Ω—è (Pruning)

### –ü—Ä–æ–±–ª–µ–º–∞ –ø–µ—Ä–µ–Ω–∞–≤—á–∞–Ω–Ω—è
–î–µ—Ä–µ–≤–∞ —Ä—ñ—à–µ–Ω—å –º–æ–∂—É—Ç—å –ª–µ–≥–∫–æ **–ø–µ—Ä–µ–Ω–∞–≤—á–∞—Ç–∏—Å—å** - –∑–∞–ø–∞–º º—è—Ç–æ–≤—É–≤–∞—Ç–∏ –Ω–∞–≤—á–∞–ª—å–Ω—ñ –¥–∞–Ω—ñ, —Å—Ç–≤–æ—Ä—é—é—á–∏ –Ω–∞–¥—Ç–æ —Å–∫–ª–∞–¥–Ω—ñ –¥–µ—Ä–µ–≤–∞, —è–∫—ñ –Ω–µ —É–∑–∞–≥–∞–ª—å–Ω—é—é—Ç—å—Å—è –¥–æ–±—Ä–µ –¥–æ –Ω–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö.

**–û–∑–Ω–∞–∫–∏ –ø–µ—Ä–µ–Ω–∞–≤—á–∞–Ω–Ω—è:**
- –î—É–∂–µ –≥–ª–∏–±–æ–∫—ñ –¥–µ—Ä–µ–≤–∞ –∑ –±–∞–≥–∞—Ç—å–º–∞ —Ä—ñ–≤–Ω—è–º–∏
- –ë–∞–≥–∞—Ç–æ –ª–∏—Å—Ç–æ–≤–∏—Ö –≤—É–∑–ª—ñ–≤ –∑ –¥—É–∂–µ –º–∞–ª–æ—é –∫—ñ–ª—å–∫—ñ—Å—Ç—é –∑—Ä–∞–∑–∫—ñ–≤
- –Ü–¥–µ–∞–ª—å–Ω—ñ –º–µ—Ç—Ä–∏–∫–∏ –Ω–∞ –Ω–∞–≤—á–∞–ª—å–Ω–∏—Ö –¥–∞–Ω–∏—Ö, –∞–ª–µ –Ω–∏–∑—å–∫—ñ –º–µ—Ç—Ä–∏–∫–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö
- –î–µ—Ä–µ–≤–∞, —è–∫—ñ –∑–∞–ø–∞–º'—è—Ç–æ–≤—É—é—Ç—å –Ω–∞–≤—á–∞–ª—å–Ω—ñ –ø—Ä–∏–∫–ª–∞–¥–∏, –∞ –Ω–µ –≤–∏–≤—á–∞—é—Ç—å –ø–∞—Ç–µ—Ä–Ω–∏

### –í—ñ–¥—Å—ñ–∫–∞–Ω–Ω—è: –†—ñ—à–µ–Ω–Ω—è –ø—Ä–æ–±–ª–µ–º–∏ –ø–µ—Ä–µ–Ω–∞–≤—á–∞–Ω–Ω—è
–í—ñ–¥—Å—ñ–∫–∞–Ω–Ω—è - —Ü–µ –ø—Ä–æ—Ü–µ—Å –≤–∏–¥–∞–ª–µ–Ω–Ω—è –≥—ñ–ª–æ–∫ –∑ –¥–µ—Ä–µ–≤–∞ —Ä—ñ—à–µ–Ω—å –¥–ª—è –ø–æ–∫—Ä–∞—â–µ–Ω–Ω—è —É–∑–∞–≥–∞–ª—å–Ω–µ–Ω–Ω—è.

#### –ü–æ–ø–µ—Ä–µ–¥–Ω—î –í—ñ–¥—Å—ñ–∫–∞–Ω–Ω—è (–†–∞–Ω–Ω—è –∑—É–ø–∏–Ω–∫–∞)
–ó—É–ø–∏–Ω–∏—Ç–∏ –ø–æ–±—É–¥–æ–≤—É –¥–µ—Ä–µ–≤–∞ –¥–æ —Ç–æ–≥–æ, —è–∫ –≤–æ–Ω–æ —Å—Ç–∞–Ω–µ –Ω–∞–¥—Ç–æ —Å–∫–ª–∞–¥–Ω–∏–º:
- –í—Å—Ç–∞–Ω–æ–≤–∏—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É –≥–ª–∏–±–∏–Ω—É
- –í—Å—Ç–∞–Ω–æ–≤–∏—Ç–∏ –º—ñ–Ω—ñ–º–∞–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∑—Ä–∞–∑–∫—ñ–≤ –Ω–∞ –ª–∏—Å—Ç

## 7. –ü—Ä–∞–∫—Ç–∏—á–Ω–∞ —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –∑ Scikit-Learn

–¢–µ–ø–µ—Ä –¥–∞–≤–∞–π—Ç–µ —Ä–µ–∞–ª—ñ–∑—É—î–º–æ –¥–µ—Ä–µ–≤–∞ —Ä—ñ—à–µ–Ω—å, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ scikit-learn –Ω–∞ —Ä–µ–∞–ª—å–Ω–æ–º—É –Ω–∞–±–æ—Ä—ñ –¥–∞–Ω–∏—Ö! –ú–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—î–º–æ **–Ω–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö Iris**, —â–æ–± –ø—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä—É–≤–∞—Ç–∏, —è–∫ –ø—Ä–∞—Ü—é—é—Ç—å –¥–µ—Ä–µ–≤–∞ —Ä—ñ—à–µ–Ω—å –Ω–∞ –ø—Ä–∞–∫—Ç–∏—Ü—ñ.

### –ù–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö Iris
–ù–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö Iris –º—ñ—Å—Ç–∏—Ç—å –≤–∏–º—ñ—Ä–∏ –∫–≤—ñ—Ç–æ–∫ —ñ—Ä–∏—Å—ñ–≤ –∑ —Ç—Ä—å–æ–º–∞ –≤–∏–¥–∞–º–∏:
- **Setosa**: 50 –∑—Ä–∞–∑–∫—ñ–≤
- **Versicolor**: 50 –∑—Ä–∞–∑–∫—ñ–≤  
- **Virginica**: 50 –∑—Ä–∞–∑–∫—ñ–≤

**–û–∑–Ω–∞–∫–∏:**
- –î–æ–≤–∂–∏–Ω–∞ —á–∞—à–æ–ª–∏—Å—Ç–∫–∞ (—Å–º)
- –®–∏—Ä–∏–Ω–∞ —á–∞—à–æ–ª–∏—Å—Ç–∫–∞ (—Å–º)
- –î–æ–≤–∂–∏–Ω–∞ –ø–µ–ª—é—Å—Ç–∫–∏ (—Å–º)
- –®–∏—Ä–∏–Ω–∞ –ø–µ–ª—é—Å—Ç–∫–∏ (—Å–º)

````python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# Load the Iris dataset
print("=== Loading Iris Dataset ===")
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names
target_names = iris.target_names

print(f"Dataset shape: {X.shape}")
print(f"Features: {feature_names}")
print(f" Classes: {target_names}")
print(f"Class distribution: {np.bincount(y)}")
print()

# Create a DataFrame for easier visualization
iris_df = pd.DataFrame(X, columns=feature_names)
iris_df['species'] = [target_names[i] for i in y]

print("First few rows of the dataset:")
print(iris_df.head())
print()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Data split:")
print(f"   Training set: {X_train.shape[0]} samples")
print(f"   Testing set:  {X_test.shape[0]} samples")
print()

# Create and train a decision tree classifier
print("üå≥ Training Decision Tree...")
dt_classifier = DecisionTreeClassifier(
    criterion='gini',        # Use Gini impurity
    max_depth=3,            # Limit tree depth to prevent overfitting
    min_samples_split=5,    # Minimum samples to split a node
    min_samples_leaf=2,     # Minimum samples in a leaf
    random_state=42
)

# Train the model
dt_classifier.fit(X_train, y_train)

# Make predictions
y_pred = dt_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy:.3f}")
print()

# Display detailed classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=target_names))
````

````python
# Let's visualize the decision tree structure!

plt.figure(figsize=(15, 10))
plot_tree(dt_classifier, 
          feature_names=feature_names,
          class_names=target_names,
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Decision Tree for Iris Classification", fontsize=16, fontweight='bold')
plt.show()

# Let's also extract the tree as text rules
print("Decision Tree Rules (Text Format):")
print("=" * 50)
tree_rules = export_text(dt_classifier, 
                        feature_names=feature_names,
                        class_names=target_names,
                        show_weights=True)
print(tree_rules)

# Let's analyze feature importance
print("\nFeature Importance Analysis:")
print("-" * 40)
feature_importance = dt_classifier.feature_importances_
for i, (feature, importance) in enumerate(zip(feature_names, feature_importance)):
    print(f"{feature:20}: {importance:.3f}")

# Create a bar plot of feature importance
plt.figure(figsize=(10, 6))
plt.barh(feature_names, feature_importance)
plt.xlabel('Feature Importance')
plt.title('Feature Importance in Decision Tree')
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nMost important feature: {feature_names[np.argmax(feature_importance)]}")
````

````python
# Let's create some additional visualizations to understand our decision tree better!

# Confusion Matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix - Decision Tree Classification')
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.show()

# Feature importance comparison
plt.figure(figsize=(12, 5))

# Subplot 1: Bar chart
plt.subplot(1, 2, 1)
bars = plt.bar(range(len(feature_names)), feature_importance, color='skyblue', edgecolor='navy')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance (Bar Chart)')
plt.xticks(range(len(feature_names)), feature_names, rotation=45)
plt.grid(True, alpha=0.3)

# Add value labels on bars
for bar, importance in zip(bars, feature_importance):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{importance:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Let's demonstrate how to make predictions on new data
print("Making predictions on new samples:")
print("-" * 50)

# Create some example new samples
new_samples = np.array([
    [5.1, 3.5, 1.4, 0.2],  # Should be setosa
    [6.2, 2.9, 4.3, 1.3],  # Should be versicolor  
    [7.2, 3.0, 5.8, 1.6]   # Should be virginica
])

predictions = dt_classifier.predict(new_samples)
probabilities = dt_classifier.predict_proba(new_samples)

for i, (sample, pred, prob) in enumerate(zip(new_samples, predictions, probabilities)):
    print(f"Sample {i+1}: {sample}")
    print(f"  Predicted: {target_names[pred]}")
    print(f"  Confidence: {prob[pred]:.3f}")
    print(f"  All probabilities: {dict(zip(target_names, prob))}")
    print()
````
