# Зменшення розмірності в data mining

## 1. Вступ до зменшення розмірності
### Визначення
**Зменшення розмірності** - це техніка в data mining та машинному навчанні, яка зменшує кількість змінних (ознак), що розглядаються. Воно перетворює дані з високовимірного простору в простір меншої розмірності, зберігаючи якомога більше значущої інформації.

![Dimensionality Reduction](../images/4.0-dim-red.png)

[Джерело зображення](https://www.geeksforgeeks.org/data-science/dimensionality-reduction-techniques/)

### Ключові концепції
- **Високовимірні дані**: Набори даних з багатьма ознаками (вимірами) відносно кількості спостережень
- **Простір ознак**: n-вимірний простір, де кожен вимір представляє ознаку
- **Внутрішня розмірність**: Мінімальна кількість вимірів, необхідна для представлення даних без значної втрати інформації
- **Проекція**: Процес відображення високовимірних даних у простір меншої розмірності

Дано набір даних X ∈ ℝ^(n×d), де:
- n = кількість спостережень
- d = кількість ознак (вимірів)

Зменшення розмірності прагне знайти перетворення T таке, що:
- T: ℝ^d → ℝ^k, де k << d
- Перетворені дані Y = T(X) зберігають основну структуру X

## 2. Прокляття розмірності
### Що таке прокляття розмірності?
"Прокляття розмірності" відноситься до різних явищ, що виникають при аналізі та організації даних у високовимірних просторах і не виникають у просторах низької розмірності. Зі збільшенням кількості вимірів об'єм простору зростає експоненційно, роблячи дані все більш розрідженими.
### Ключові проблеми
#### 1. **Розрідженість даних**
- У високих вимірах більшість точок даних стають рівновіддаленими одна від одної
- Концепція "найближчого сусіда" втрачає сенс
- Дані стають все більш розрідженими, що ускладнює розпізнавання патернів
#### 2. **Обчислювальна складність**
- Багато алгоритмів мають обчислювальну складність, що зростає експоненційно з розмірністю
- Вимоги до зберігання зростають експоненційно
- Час обробки стає непрактичним
#### 3. **Перенавчання**
- Моделі схильні до перенавчання у високовимірних просторах
- Кількість параметрів швидко зростає з розмірністю
- Здатність до узагальнення погіршується
#### 4. **Концентрація відстаней**
- Всі попарні відстані стають подібними у високих вимірах
- Кластеризація та класифікація стають складними
### Проблема візуалізації
- Люди можуть безпосередньо візуалізувати лише до 3 вимірів
- Високовимірні дані вимагають технік проекції або вбудовування
- Втрата просторової інтуїції у високовимірних просторах

## 3. Коли використовувати зменшення розмірності
### Основні випадки використання
#### 1. **Візуалізація даних**
- **Коли**: У вас більше 3 ознак і ви хочете візуалізувати патерни
- **Мета**: Створити 2D або 3D графіки, що розкривають структуру даних
- **Методи**: МГК (PCA), t-SNE, UMAP, MDS
#### 2. **Обчислювальна ефективність**
- **Коли**: Високовимірні дані спричиняють проблеми з продуктивністю
- **Мета**: Зменшити час обчислень та використання пам'яті
- **Методи**: МГК(PCA), Випадкові проекції, Вибір ознак
#### 3. **Запобігання перенавчанню**
- **Коли**: Кількість ознак наближається або перевищує кількість зразків
- **Мета**: Покращити узагальнення моделі
- **Методи**: МГК (PCA), LDA, Техніки регуляризації
#### 4. **Зменшення шуму**
- **Коли**: Дані містять нерелевантні або зашумлені ознаки
- **Мета**: Зосередитися на найбільш інформативних ознаках
- **Методи**: МГК (PCA), ICA, Факторний аналіз
### Коли НЕ використовувати
- **Інтерпретованість критична**: Зменшені виміри можуть бути неінтерпретованими
- **Всі ознаки однаково важливі**: Немає чіткої збитковості
- **Малі набори даних**: Може втратитися важлива інформація
- **Лише категоріальні дані**: Більшість методів краще працюють з неперервними даними

## 4. Типи зменшення розмірності
### Лінійні методи
#### **Метод головних компонент, МГК (Principal Component Analysis, PCA)**
- **Принцип**: Знаходить ортогональні напрямки максимальної дисперсії
- **Випадок використання**: Загальне лінійне зменшення розмірності
- **Переваги**: Швидкий, інтерпретований, зберігає глобальну структуру
- **Обмеження**: Припускає лінійні зв'язки
#### **Лінійний дискримінантний аналіз, ЛДА (Linear Discriminant Analysis, LDA)**
- **Принцип**: Максимізує розділення між класами, мінімізуючи дисперсію всередині класів
- **Випадок використання**: Зменшення розмірності для класифікації
- **Переваги**: Використовує мітки класів, корисно для класифікації
- **Обмеження**: Обмежений лінійними межами, вимагає розмічених даних
### Нелінійні методи
#### **t-розподілене стохастичне вбудовування сусідів (t-SNE)**
- **Принцип**: Зберігає локальну структуру сусідства, використовуючи ймовірнісні розподіли
- **Випадок використання**: Візуалізація високовимірних даних
- **Переваги**: Відмінний для візуалізації, зберігає локальну структуру
- **Обмеження**: Обчислювально дорогий
#### **Рівномірне наближення та проекція многовиду (UMAP)**
- **Принцип**: Зберігає як локальну, так і глобальну структуру, використовуючи апроксимацію многовиду
- **Випадок використання**: Загальне нелінійне зменшення розмірності
- **Переваги**: Швидкий, зберігає як локальну, так і глобальну структуру
- **Обмеження**: Чутливий до гіперпараметрів

### Метрики успіху
#### **Кількісні міри**
- **Пояснена дисперсія**: Відсоток оригінальної дисперсії, що зберігається
- **Помилка реконструкції**: Наскільки добре оригінальні дані можуть бути реконструйовані
- **Точність класифікації**: Продуктивність на подальших завданнях
- **Обчислювальна ефективність**: Швидкість та використання пам'яті
#### **Якісні міри**
- **Інтерпретованість**: Наскільки добре зменшені виміри можуть бути зрозумілі
- **Якість візуалізації**: Чіткість 2D/3D проекцій
- **Валідація експертом предметної області**: Експертна оцінка результатів
- **Бізнес-вплив**: Покращення бізнес-метрик

## 5. Метод головних компонент, МГК (Principal Component Analysis, PCA)
### Інтуїтивне розуміння
Уявіть, що у вас є набір даних про людей з ознаками (вимірами), такими як зріст, вага, вік та розмір взуття. Ви помічаєте, що вищі люди, як правило, важчі, і є сильна кореляція між цими змінними. МГК запитує: "Що якщо ми могли б описати ці дані, використовуючи менше вимірів, але більш значущих?"
Замість використання зросту та ваги окремо, МГК може знайти новий вимір під назвою "розмір тіла", який поєднує інформацію про зріст та вагу. Цей новий вимір захоплює більшість варіації в оригінальних даних, використовуючи менше ознак.
### Основна ідея
МГК знаходить напрямки (головні компоненти), у яких дані варіюються найбільше. Ці напрямки:
1. **Ортогональні** (перпендикулярні) один одному
2. **Впорядковані** за кількістю дисперсії, яку вони пояснюють
3. **Лінійні комбінації** оригінальних ознак
### Геометрична інтуїція
Подумайте про ваші дані як про хмару точок у просторі. МГК знаходить:
- **Перша головна компонента**: Напрямок, уздовж якого дані варіюються найбільше
- **Друга головна компонента**: Напрямок максимальної залишкової дисперсії, перпендикулярний до першої
- **Третя головна компонента**: Напрямок максимальної залишкової дисперсії, перпендикулярний до обох попередніх компонент
- І так далі...
### Математичні основи
#### Крок 1: Центрування даних
Спочатку ми центруємо дані, віднімаючи середнє:
$$
X_{\text{centered}} = X - \mu
$$
де $\mu$ - вектор середніх значень всіх ознак.
#### Крок 2: Коваріаційна матриця
Ми обчислюємо коваріаційну матрицю $C$:
$$
C = \frac{1}{n} X_{\text{centered}}^T X_{\text{centered}}
$$
Коваріаційна матриця показує, як ознаки варіюються разом.
#### Крок 3: Знаходження власних чисел та власних векторів
Ми знаходимо власні значення та власні вектори коваріаційної матриці:
$$
C \, v_i = \lambda_i \, v_i
$$
де:
- $\lambda_i$ - власні значення (дисперсії)
- $v_i$ - власні вектори (головні компоненти)
#### Крок 4: Головні компоненти
Власні вектори $v_i$ є головними компонентами, впорядкованими за відповідними власними значеннями (найбільші спочатку).
### Алгоритм покроково
1. **Стандартизувати дані** (опціонально, але рекомендується)
2. **Центрувати дані**, віднімаючи середнє
3. **Обчислити коваріаційну матрицю**
4. **Знайти власні значення та власні вектори** коваріаційної матриці
5. **Відсортувати власні вектори** за власними значеннями в порядку спадання
6. **Вибрати топ k власних векторів** для зменшення розмірності
7. **Перетворити дані**, використовуючи вибрані власні вектори

### Математична формулювання
Дано матрицю даних $X \in \mathbb{R}^{n \times d}$:

1. **Центрування**:
$$
\tilde{X} = X - \mathbf{1}_n \mu^T
$$
де $\mu$ - вектор середніх значень стовпців $X$, а $\mathbf{1}_n$ - вектор-стовпець одиниць довжини $n$.

2. **Коваріаційна матриця**:
$$
C = \frac{1}{n} \tilde{X}^T \tilde{X}
$$

3. **Знаходження власних чисел та власних векторів**:
$$
C = V \Lambda V^T
$$
де $V$ містить власні вектори, а $\Lambda$ - діагональна матриця власних значень.

4. **Головні компоненти**:
$$
W = V_{[:, 1:k]}
$$
(перші $k$ стовпців $V$)

5. **Перетворені дані**:
$$
Y = \tilde{X} W
$$

### Ключові властивості
#### **Ортогональність**
Головні компоненти ортогональні одна одній:
$$
\mathbf{v}_i^T \mathbf{v}_j = 0 \quad \text{for } i \neq j
$$
#### **Максимізація дисперсії**
Кожна головна компонента захоплює максимально можливу дисперсію, що залишилася після попередніх компонент:
$$
\mathrm{Var}(\mathrm{PC}_i) = \lambda_i
$$
#### **Некорельовані компоненти**
Перетворені ознаки некорельовані:
$$
\mathrm{Cov}(\mathrm{PC}_i, \mathrm{PC}_j) = 0 \quad \text{for } i \neq j
$$
### Пояснена дисперсія
Частка дисперсії, пояснена $i$-ю головною компонентою:
$$
\text{Variance Explained}_i = \frac{\lambda_i}{\sum_j \lambda_j}
$$
Кумулятивна дисперсія, пояснена першими $k$ компонентами:
$$
\text{Cumulative Variance} = \frac{\sum_{i=1}^k \lambda_i}{\sum_j \lambda_j}
$$
### Помилка реконструкції
Помилка реконструкції вимірює, наскільки добре оригінальні дані можуть бути реконструйовані з представлення зменшеної розмірності. Після проекції даних на перші $k$ головних компонент, ми можемо спробувати реконструювати оригінальні дані, проекціюючи назад у простір оригінальних ознак.

Дано перетворені дані $Y = XW_k$, де $W_k$ містить перші $k$ головних компонент, реконструйовані дані:
$$
\hat{X} = Y W_k^T = X W_k W_k^T
$$

Помилка реконструкції для однієї точки даних $\mathbf{x}_i$ - це квадратна евклідова відстань між оригінальною та реконструйованою точкою:
$$
\text{Reconstruction Error}_i = ||\mathbf{x}_i - \hat{\mathbf{x}}_i||^2
$$

Середня квадратична помилка реконструкції (СКПР) для всіх $n$ точок даних:
$$
\text{MSRE} = \frac{1}{n} \sum_{i=1}^{n} ||\mathbf{x}_i - \hat{\mathbf{x}}_i||^2 = \frac{1}{n} \sum_{i=1}^{n} ||\mathbf{x}_i - X W_k W_k^T||^2
$$

Ключова властивість МГК полягає в тому, що помилка реконструкції дорівнює сумі власних значень, що відповідають відкинутим головним компонентам:
$$
\text{MSRE} = \sum_{j=k+1}^{d} \lambda_j
$$

Це означає, що мінімізація помилки реконструкції еквівалентна максимізації дисперсії, поясненої збереженими компонентами. Нижча помилка реконструкції вказує на краще збереження оригінальної структури даних.
### Вибір кількості компонент
#### **Метод Scree Plot**
- Побудувати власні значення в порядку спадання
- Шукати точку "ліктя", де крива вирівнюється
- Вибрати компоненти до точки ліктя

#### **Поріг дисперсії**
- Вибрати мінімальну кількість компонент, що пояснюють певний відсоток дисперсії (наприклад, 95%)
- Обчислити кумулятивну пояснену дисперсію
#### **Критерій Кайзера**
- Зберегти компоненти з власними значеннями > 1 (для стандартизованих даних)
- Заснований на ідеї, що кожна компонента повинна пояснювати принаймні стільки ж дисперсії, скільки одна оригінальна змінна
### Переваги МГК
1. **Лінійне перетворення**: Зберігає лінійні зв'язки
2. **Некорельовані ознаки**: Зменшує мультиколінеарність
3. **Збереження дисперсії**: Максимізує збереження інформації
4. **Обчислювальна ефективність**: Швидкий та масштабований
5. **Інтерпретованість**: Компоненти є лінійними комбінаціями оригінальних ознак
### Обмеження МГК
1. **Лінійне припущення**: Не може захопити нелінійні зв'язки
2. **Чутливість до масштабування**: Вимагає правильної попередньої обробки даних
3. **Глобальна оптимізація**: Може не зберігати локальну структуру
4. **Інтерпретованість**: Головні компоненти можуть не мати чіткого значення
5. **Чутливість до викидів**: Екстремальні значення можуть значно вплинути на результати
### Коли МГК працює найкраще
- **Лінійні зв'язки** між ознаками
- **Корельовані ознаки**, які можна поєднати
- **Гаусівські або майже гаусівські** розподіли даних
- **Великі набори даних** з багатьма ознаками
- **Зменшення шуму** необхідне
### Коли МГК може працювати не дуже добре
- **Нелінійні зв'язки** між ознаками
- **Категоріальні дані** (використовувати інші методи)
- **Розріджені дані** (розглянути розріджений МГК)
- **Малі набори даних** відносно розмірності
- **Інтерпретованість** критична (компоненти можуть бути важкими для інтерпретації)

## 6. МГК на практиці: Приклад з набором даних Iris
Застосуємо МГК до набору даних Iris, щоб побачити, як він працює на практиці. Набір даних Iris містить виміри 150 квіток ірисів трьох різних видів, з 4 ознаками кожна.

- **150 зразків** (50 на вид)
- **4 ознаки**: довжина чашолистка, ширина чашолистка, довжина пелюстки, ширина пелюстки
- **3 класи**: Setosa, Versicolor, Virginica
- **Мета**: Зменшити з 4D до 2D, зберігаючи розділення видів

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

print("Dataset shape:", X.shape)
print("Features:", feature_names)
print("Classes:", target_names)
print("\nFirst 5 samples:")
print(pd.DataFrame(X, columns=feature_names).head())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for i, feature in enumerate(feature_names):
    axes[i].hist(X[y==0, i], alpha=0.7, label='Setosa', bins=15)
    axes[i].hist(X[y==1, i], alpha=0.7, label='Versicolor', bins=15)
    axes[i].hist(X[y==2, i], alpha=0.7, label='Virginica', bins=15)
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.tight_layout()
plt.show()

# Check correlations
corr_matrix = np.corrcoef(X.T)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, xticklabels=feature_names, 
            yticklabels=feature_names, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```

### Крок 1: Попередня обробка даних
Перед застосуванням МГК нам потрібно стандартизувати наші дані. Це критично важливо, оскільки МГК чутливий до масштабу ознак.

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Original data statistics:")
print(f"Mean: {X.mean(axis=0)}")
print(f"Std:  {X.std(axis=0)}")

print("\nStandardized data statistics:")
print(f"Mean: {X_scaled.mean(axis=0)}")
print(f"Std:  {X_scaled.std(axis=0)}")

# Show first few standardized samples
print("\nFirst 5 standardized samples:")
print(pd.DataFrame(X_scaled, columns=feature_names).head())
```

### Крок 2: Застосувати МГК
Тепер застосуємо МГК до наших стандартизованих даних та дослідимо результати.

```python
# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Get explained variance ratios
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

print("Explained variance by each component:")
for i, (var, cum_var) in enumerate(zip(explained_variance_ratio, cumulative_variance)):
    print(f"PC{i+1}: {var:.3f} ({var*100:.1f}%) - Cumulative: {cum_var:.3f} ({cum_var*100:.1f}%)")

print(f"\nTotal variance explained by first 2 components: {cumulative_variance[1]:.3f} ({cumulative_variance[1]*100:.1f}%)")
```

```python
# Visualize explained variance
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Scree plot
ax1.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, 
        alpha=0.7, color='skyblue', edgecolor='navy')
ax1.set_xlabel('Principal Component')
ax1.set_ylabel('Explained Variance Ratio')
ax1.set_title('Scree Plot - Individual Variance Explained')
ax1.grid(True, alpha=0.3)

# Cumulative variance plot
ax2.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 
         'bo-', linewidth=2, markersize=8)
ax2.axhline(y=0.95, color='r', linestyle='--', alpha=0.7, label='95% threshold')
ax2.set_xlabel('Number of Components')
ax2.set_ylabel('Cumulative Explained Variance')
ax2.set_title('Cumulative Variance Explained')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()
```

Подивимося, як наші дані виглядають у зменшеному 2D просторі та порівняємо їх з оригінальними 4D даними.

```python
# Examine the principal components (loadings)
components = pca.components_
feature_names = iris.feature_names

print("Principal Component Loadings (how original features contribute to PCs):")
print("=" * 60)
for i, pc in enumerate(components):
    print(f"\nPC{i+1} (explains {explained_variance_ratio[i]:.1%} of variance):")
    for j, (feature, loading) in enumerate(zip(feature_names, pc)):
        print(f"  {feature:15s}: {loading:6.3f}")

# Visualize loadings
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# PC1 loadings
bars1 = ax1.bar(feature_names, components[0], color='skyblue', alpha=0.7)
ax1.set_title(f'PC1 Loadings (explains {explained_variance_ratio[0]:.1%} variance)')
ax1.set_ylabel('Loading Value')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(True, alpha=0.3)

# PC2 loadings
bars2 = ax2.bar(feature_names, components[1], color='lightcoral', alpha=0.7)
ax2.set_title(f'PC2 Loadings (explains {explained_variance_ratio[1]:.1%} variance)')
ax2.set_ylabel('Loading Value')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. t-розподілене стохастичне вбудовування сусідів (t-SNE)
### Інтуїтивне розуміння
Уявіть, що у вас є високовимірний набір даних, де точки даних утворюють кластери та складні структури, які неможливо візуалізувати безпосередньо. t-SNE відповідає на питання: "Як ми можемо створити 2D або 3D карту, яка зберігає локальні зв'язки у наших даних?"
Подумайте про це як про створення карти міста на основі того, як люди переміщуються між районами. Якщо два райони часто відвідуються одними й тими ж людьми, вони повинні бути близькими на карті, навіть якщо вони далеко один від одного географічно.
### Основна ідея
t-SNE зосереджується на **збереженні локальної структури** замість глобальної структури. Він:
1. **Вимірює схожість** між точками даних у високовимірному просторі
2. **Створює ймовірнісний розподіл**, що відображає цю схожість
3. **Відображає у низькі виміри**, зберігаючи структуру ймовірності
4. **Використовує t-розподіл** для обробки "проблеми скупчення" у низьких вимірах
### Ключові відмінності від МГК
| Аспект | МГК | t-SNE |
|--------|-----|-------|
| **Збережена структура** | Глобальна (дисперсія) | Локальна (сусідства) |
| **Лінійний/Нелінійний** | Лінійний | Нелінійний |
| **Збереження відстаней** | Евклідові відстані | Ймовірнісні розподіли |
| **Обчислювальна вартість** | Швидкий | Повільний |
| **Нові дані** | Легко перетворити | Вимагає повторного навчання |
| **Інтерпретованість** | Компоненти інтерпретовані | Лише візуальні кластери |
### Математичні основи
#### Крок 1: Схожість у високих вимірах
Для кожної пари точок (i, j) обчислити схожість, використовуючи гаусівське ядро:
$$
p_{j|i} = \frac{\exp\left(-\frac{||x_i - x_j||^2}{2\sigma_i^2}\right)}{\sum_{k \neq i} \exp\left(-\frac{||x_i - x_k||^2}{2\sigma_i^2}\right)}
$$
Де $\sigma_i$ вибирається так, щоб перплексність розподілу дорівнювала попередньо визначеному значенню.
#### Крок 2: Симетрична схожість
Зробити розподіл симетричним:
$$
p_{ij} = \frac{p_{j|i} + p_{i|j}}{2n}
$$
#### Крок 3: Схожість у низьких вимірах
У просторі низької розмірності використовувати t-розподіл (з 1 ступенем свободи):
$$
q_{ij} = \frac{(1 + ||y_i - y_j||^2)^{-1}}{\sum_{k \neq l} (1 + ||y_k - y_l||^2)^{-1}}
$$
#### Крок 4: Оптимізація
Мінімізувати розбіжність Кульбака-Лейблера:
$$
KL(P||Q) = \sum_{i,j} p_{ij} \log \left( \frac{p_{ij}}{q_{ij}} \right)
$$
#### Алгоритм покроково
1. **Встановити параметри**: перплексність, швидкість навчання, кількість ітерацій
2. **Обчислити схожість у високих вимірах**: Створити ймовірнісний розподіл P
3. **Ініціалізувати точки у низьких вимірах**: Випадкові позиції в цільовому просторі
4. **Ітеративна оптимізація**: 
   - Обчислити схожість у низьких вимірах Q
   - Обчислити градієнт розбіжності KL
   - Оновити позиції, використовуючи градієнтний спуск
5. **Повторювати** до збіжності або максимальної кількості ітерацій
### Ключові параметри
#### **Перплексність**
- **Діапазон**: 5-50 (зазвичай 30)
- **Ефект**: Контролює розмір локальних сусідств
- **Низька перплексність**: Зосереджується на дуже локальній структурі
- **Висока перплексність**: Враховує більш глобальну структуру
#### **Швидкість навчання**
- **Типовий діапазон**: 10-1000
- **Ефект**: Контролює розмір кроку в оптимізації
- **Занадто низька**: Повільна збіжність, може застрягнути
- **Занадто висока**: Нестабільна, може розходитися
#### **Кількість ітерацій**
- **Типовий діапазон**: 250-1000
- **Ефект**: Більше ітерацій = краща збіжність
- **Компроміс**: Час обчислень проти якості
### Проблема скупчення
У високих вимірах є більше "місця" для точок, щоб бути далеко одна від одної. При відображенні у низькі виміри точки "скупчуються" разом. t-SNE використовує t-розподіл для вирішення цього:
- **Гаусівський розподіл**: Важкі хвости у високих вимірах, легкі хвости у низьких вимірах
- **t-розподіл**: Важкі хвости як у високих, так і у низьких вимірах
- **Результат**: Краще збереження локальної структури
### Переваги t-SNE
1. **Відмінна візуалізація**: Створює красиві, інтерпретовані графіки
2. **Нелінійний**: Захоплює складні, вигнуті зв'язки
3. **Локальна структура**: Зберігає зв'язки сусідства
4. **Кластеризація**: Виявляє природні групи в даних
5. **Надійність**: Менш чутливий до викидів, ніж МГК
### Обмеження t-SNE
1. **Обчислювальна вартість**: Повільний для великих наборів даних (>10,000 точок)
2. **Немає нових даних**: Неможливо перетворити нові точки без повторного навчання
3. **Глобальна структура**: Може не зберігати відстані між віддаленими кластерами
4. **Чутливість до параметрів**: Результати сильно залежать від перплексності
5. **Недетермінований**: Різні запуски можуть давати різні результати
6. **Інтерпретованість**: Осі низької розмірності не мають чіткого значення
### Коли t-SNE працює найкраще
- **Візуалізація**: Створення 2D/3D графіків для дослідження
- **Відкриття кластерів**: Знаходження природних груп у даних
- **Нелінійні дані**: Складні, вигнуті зв'язки
- **Малі та середні набори даних**: <10,000 точок даних
- **Локальна структура**: Коли локальні зв'язки важливі
### Коли t-SNE може працювати не дуже добре
- **Великі набори даних**: Обчислювально дорого
- **Глобальна структура**: Коли відстані між кластерами мають значення
- **Нові точки даних**: Неможливо легко додати нові точки
- **Лінійні зв'язки**: МГК може бути більш доречним
- **Відтворюваність**: Потрібно фіксувати random seed для відтворюваних результатів

## 8. t-SNE на практиці: Приклад з набором даних Iris
Тепер застосуємо t-SNE до того ж набору даних Iris, щоб побачити, як він порівнюється з МГК. Ми дослідимо різні значення перплексності та побачимо, як t-SNE розкриває локальну структуру, яку МГК може пропустити.
### Огляд набору даних
- **Той самий набір даних Iris**: 150 зразків, 4 ознаки, 3 види
- **Мета**: Порівняти візуалізацію t-SNE vs МГК
- **Фокус**: Збереження локального сусідства vs глобальна дисперсія

```python
# Import t-SNE and additional libraries
from sklearn.manifold import TSNE
import time

# Set random seed for reproducibility
np.random.seed(42)
```

### Крок 1: Застосувати t-SNE з різними значеннями перплексності
Дослідимо, як різні значення перплексності впливають на візуалізацію t-SNE.

```python
# Apply t-SNE with different perplexity values
perplexities = [5, 15, 30, 50]
tsne_results = {}

print("Applying t-SNE with different perplexity values...")
print("=" * 50)

for perp in perplexities:
    print(f"Running t-SNE with perplexity={perp}...")
    start_time = time.time()
    
    tsne = TSNE(n_components=2, perplexity=perp, random_state=42, 
                learning_rate=200, max_iter=1000)
    X_tsne = tsne.fit_transform(X_scaled)
    
    end_time = time.time()
    tsne_results[perp] = X_tsne
    
    print(f"  Completed in {end_time - start_time:.2f} seconds")
    print(f"  Final KL divergence: {tsne.kl_divergence_:.3f}")

print("\nAll t-SNE runs completed!")
```

```python
# Visualize t-SNE results with different perplexities
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.ravel()

colors = ['red', 'green', 'blue']
markers = ['o', 's', '^']

for i, (perp, ax) in enumerate(zip(perplexities, axes)):
    X_tsne = tsne_results[perp]
    
    for j, (species, color, marker) in enumerate(zip(target_names, colors, markers)):
        mask = y == j
        ax.scatter(X_tsne[mask, 0], X_tsne[mask, 1], 
                  c=color, marker=marker, s=100, alpha=0.7, 
                  label=f'{species} (n={mask.sum()})')
    
    ax.set_title(f't-SNE with Perplexity={perp}')
    ax.set_xlabel('t-SNE Dimension 1')
    ax.set_ylabel('t-SNE Dimension 2')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Notice how different perplexity values create different visualizations!")
print("Lower perplexity = more local structure, Higher perplexity = more global structure")
```
